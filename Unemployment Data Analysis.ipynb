```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## 5. COVID-19 Impact Analysis
Compared unemployment before and during COVID-19...
```

```python
region_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

plt.figure()
region_avg.plot(kind='bar')
plt.title("Average Unemployment by Region")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
covid_data = df[df['Date'].dt.year == 2020]
```

```python
before_covid = df[df['Date'].dt.year < 2020]
during_covid = df[df['Date'].dt.year == 2020]

print(before_covid['Estimated Unemployment Rate (%)'].mean())
print(during_covid['Estimated Unemployment Rate (%)'].mean())
```

```python
df['Month'] = df['Date'].dt.month

monthly_avg = df.groupby('Month')['Estimated Unemployment Rate (%)'].mean()

plt.figure()
monthly_avg.plot()
plt.title("Monthly Unemployment Trend")
plt.show()
```
